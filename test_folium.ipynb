```python
import folium
import pandas as pd
import numpy as np
```

```python
coordinates = pd.read_csv('data/coords_utm_montguers.csv')
coordinates.head()
```

```python
coordinates.shape
```

```python
np.random.seed(0)
random_idxs = np.random.choice(len(coordinates), size=100, replace=False)
random_idxs
```

```python
random_coords = coordinates.iloc[random_idxs]
random_coords
```

```python
locationlist = random_coords.values.tolist()
print(len(locationlist))
print(locationlist[7])
```

```python
fmap = folium.Map(location=[44.304303, 5.506360], zoom_start=12)
for point in range(len(locationlist)):
    folium.Marker(locationlist[point]).add_to(fmap)
fmap
```
